# Faster Prime Finding

### Write a function that returns a list of all primes up to a given number.

### For each number...

### 1. Find the square root of the number
### 2. Find all the primes up to the square root
### 3. Test to see if any of those primes are devisors

### If a number has no prime divisors, it's a prime!

```python
def allPrimesUpTo(maxNum):
    primes = [2]
    if maxNum < 2:
        return None
    elif maxNum == 2:
        return primes
    else:
        for num in range(3, (maxNum+1)):
            # 1. Find square root of the number
            sqrtNum = num ** 0.5
            
            # 2. Find all the primes up to the square root
            iCount = 0
            for i in primes:
                if i > sqrtNum:
                    sqrtPrimes = primes[:iCount]
                    break
                else:
                    iCount = iCount + 1
                    
            # 3. Test to see if those primes are divisors
            for i in sqrtPrimes:
                if num % i == 0:
                    break
            else:
                primes.append(num)
                
    return primes
```

```python
def allPrimesUpTo(maxNumber):
    primes = [2]
    for number in range(3, maxNumber):
        sqrtNumber = number ** 0.5  
        for factor in primes:
            if number % factor == 0:
                break
            elif factor > sqrtNumber:
                primes.append(number)
                break              
    return primes
```
