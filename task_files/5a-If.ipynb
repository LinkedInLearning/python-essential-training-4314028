# If statements with "FizzBuzz"

```python
# 1, 2, fizz, 4, buzz, fizz, 7, 8, fizz, buzz, 11, fizz, 13, fizzbuzz, 16
# for 1:100, if /3 = fizz, /5 = buzz, /15 = fizzbuzz

for num in range(1, 101):
    if num % 15 == 0:
        print('FizzBuzz')
    else:
        if num % 3 == 0:
            print('Fizz')
        else:
            if num % 5 == 0:
                print('Buzz')
            else:
                print(num)
    
```

## Elif statements (i.e. else if statements)

```python
# 1, 2, fizz, 4, buzz, fizz, 7, 8, fizz, buzz, 11, fizz, 13, fizzbuzz, 16
# for 1:100, if /3 = fizz, /5 = buzz, /15 = fizzbuzz

# requires if statement at the start, any number of elif statements, and an optional else statement at the end

for num in range(1, 101):
    if num % 15 == 0:
        print('FizzBuzz')
    elif num % 3 == 0:
        print('Fizz')
    elif num % 5 == 0:
        print('Buzz')
    else:
        print(num)
```

## Single line

```python
n = 5
print('Fizz' if n % 3 == 0 else n)
```
